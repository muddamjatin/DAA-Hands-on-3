```python
import time
import random
import matplotlib.pyplot as plt

def is_prime_trial_division(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

def sieve_of_eratosthenes(n):
    primes = [True] * (n + 1)
    primes[0] = primes[1] = False
    p = 2
    while p * p <= n:
        if primes[p]:
            for i in range(p * p, n + 1, p):
                primes[i] = False
        p += 1
    return [i for i in range(n + 1) if primes[i]]

def is_prime_fermat(n, k=5):
    if n <= 1:
        return False
    if n <= 3:
        return True
    for _ in range(k):
        a = random.randint(2, n - 2)
        if pow(a, n - 1, n) != 1:
            return False
    return True

def is_prime_miller_rabin(n, k=5):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0:
        return False

    def miller_test(d, n):
        a = random.randint(2, n - 2)
        x = pow(a, d, n)
        if x == 1 or x == n - 1:
            return True
        while d != n - 1:
            x = (x * x) % n
            d *= 2
            if x == 1:
                return False
            if x == n - 1:
                return True
        return False

    d = n - 1
    while d % 2 == 0:
        d //= 2
    for _ in range(k):
        if not miller_test(d, n):
            return False
    return True

def is_prime_aks(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    r = 0
    d = n - 1
    while d % 2 == 0:
        r += 1
        d //= 2
    for _ in range(5):  # Number of iterations
        a = random.randint(2, n - 2)
        x = pow(a, d, n)
        if x == 1 or x == n - 1:
            continue
        for _ in range(r - 1):
            x = pow(x, 2, n)
            if x == n - 1:
                break
        else:
            return False
    return True

def benchmark_prime_testing(algorithm, n):
    start_time = time.time()
    is_prime = algorithm(n)
    end_time = time.time()
    elapsed_time = end_time - start_time
    return is_prime, elapsed_time

if __name__ == "__main__":
    num_to_test = 10**6
    num_tests = 10

    print(f"Benchmarking primality testing algorithms for {num_tests} iterations with number {num_to_test}:")

    algorithms = [
        ("Trial Division", is_prime_trial_division),
        ("Sieve of Eratosthenes", sieve_of_eratosthenes),
        ("Fermat's Test", is_prime_fermat),
        ("Miller-Rabin Test", is_prime_miller_rabin),
        ("AKS Primality Test", is_prime_aks)
    ]

    average_times = []
    for name, algorithm in algorithms:
        total_time = 0
        for _ in range(num_tests):
            _, elapsed_time = benchmark_prime_testing(algorithm, num_to_test)
            total_time += elapsed_time
        avg_time = total_time / num_tests
        print(f"{name}: Average Time = {avg_time:.6f} seconds")
        average_times.append(avg_time)

    # Plotting the results
    plt.figure(figsize=(10, 6))
    plt.bar([name for name, _ in algorithms], average_times, color='skyblue')
    plt.xlabel('Primality Testing Algorithms')
    plt.ylabel('Average Time (seconds)')
    plt.title(f'Average Time Taken by Primality Testing Algorithms ({num_to_test} iterations)')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig('primality_testing_benchmark.png')  # Save the plot as an image
    plt.show()
```
